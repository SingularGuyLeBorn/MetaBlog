## 8. ä½¿ç”¨åŸç”Ÿ HTTP è¯·æ±‚æµ‹è¯•

ä½¿ç”¨ requests åº“ç›´æ¥å‘é€ HTTP è¯·æ±‚è¿›è¡Œæµ‹è¯•ã€‚

```python
def test_raw_http_request():
    """ä½¿ç”¨åŸç”Ÿ HTTP è¯·æ±‚æµ‹è¯•"""
    print("\nğŸŒ æµ‹è¯•7: åŸç”Ÿ HTTP è¯·æ±‚")
    print("="*50)
    
    headers = {
        "Authorization": f"Bearer {API_KEY}",
        "Content-Type": "application/json"
    }
    
    payload = {
        "model": "glm-4.6v-flash",
        "messages": [
            {
                "role": "user",
                "content": [
                    {
                        "type": "video_url",
                        "video_url": {
                            "url": TEST_VIDEO_URL
                        }
                    },
                    {
                        "type": "text",
                        "text": "è¯·ç”¨ä¸€å¥è¯æ¦‚æ‹¬è¿™ä¸ªè§†é¢‘"
                    }
                ]
            }
        ]
    }
    
    try:
        response = requests.post(
            f"{BASE_URL}/chat/completions",
            headers=headers,
            json=payload,
            timeout=120
        )
        
        print(f"çŠ¶æ€ç : {response.status_code}")
        
        if response.status_code == 200:
            result = response.json()
            content = result["choices"][0]["message"]["content"]
            print("\nâœ… æˆåŠŸå“åº”:")
            display(Markdown(content))
            
            # æ˜¾ç¤ºå®Œæ•´å“åº”ç»“æ„
            print("\nğŸ“‹ å®Œæ•´å“åº”ç»“æ„:")
            print(json.dumps(result, indent=2, ensure_ascii=False)[:1000] + "...")
        else:
            print(f"âŒ è¯·æ±‚å¤±è´¥: {response.text}")
            
    except Exception as e:
        print(f"âŒ é”™è¯¯: {e}")

test_raw_http_request()
```
