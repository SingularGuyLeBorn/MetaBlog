```python
# å®šä¹‰å›¾åƒç”Ÿæˆå‡½æ•°
def generate_image(prompt, size="1024x1024", n=1, quality="standard", style="natural"):
    """
    ç”Ÿæˆå›¾åƒ
    
    Args:
        prompt: å›¾åƒæè¿°æ–‡æœ¬
        size: åˆ†è¾¨ç‡ï¼Œæ”¯æŒ 1024x1024, 768x1344, 864x1152, 1344x768, 1152x864, 1440x720, 720x1440
        n: ç”Ÿæˆæ•°é‡ï¼ˆ1-4ï¼‰
        quality: è´¨é‡ standard æˆ– hd
        style: é£æ ¼ natural æˆ– vivid
    
    Returns:
        ç”Ÿæˆç»“æœ
    """
    try:
        response = client.images.generate(
            model="cogview-3-flash",
            prompt=prompt,
            size=size,
            n=n,
            quality=quality,
            style=style
        )
        return response
    except Exception as e:
        print(f"âŒ ç”Ÿæˆå¤±è´¥ï¼š{e}")
        return None

# å®šä¹‰æ˜¾ç¤ºå›¾åƒå‡½æ•°
def display_image(url, width=400):
    """æ˜¾ç¤ºå›¾åƒ"""
    try:
        display(IPImage(url=url, width=width))
    except Exception as e:
        print(f"æ— æ³•æ˜¾ç¤ºå›¾åƒï¼š{e}")
        print(f"å›¾åƒURLï¼š{url}")

print("âœ… è¾…åŠ©å‡½æ•°å·²å®šä¹‰")
```
