# 07-FIM 补全（Beta）- DeepSeek API

⚠️ 注意：
- 需要使用 `https://api.deepseek.com/beta` 端点
- 仅 `deepseek-chat` 支持，`deepseek-reasoner` 不支持

```python
from openai import OpenAI
import os
from dotenv import load_dotenv

load_dotenv(dotenv_path='../../.env')

api_key = os.getenv("VITE_DEEPSEEK_API_KEY")
if not api_key:
    raise ValueError("❌ 未找到 API Key")

# ⚠️ 使用 beta 端点
client = OpenAI(
    api_key=api_key,
    base_url="https://api.deepseek.com/beta",
)
```

## 基础 FIM 补全

```python
# FIM 补全（使用 /completions 端点，非 /chat/completions）
response = client.completions.create(
    model="deepseek-chat",  # 仅支持 deepseek-chat
    prompt="def fib(a):",
    suffix="    return fib(a-1) + fib(a-2)",
    max_tokens=128
)

print(response.choices[0].text)
```

## 函数补全示例

```python
code_prefix = '''def fibonacci(n):
    if n <= 1:
        return n
    '''

code_suffix = '''
    return result'''

response = client.completions.create(
    model="deepseek-chat",
    prompt=code_prefix,
    suffix=code_suffix,
    max_tokens=256,
    temperature=0.2
)

completed_code = code_prefix + response.choices[0].text + code_suffix
print(completed_code)
```
