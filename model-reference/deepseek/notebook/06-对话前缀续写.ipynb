# 06-对话前缀续写（Beta）- DeepSeek API

⚠️ 需要使用 `https://api.deepseek.com/beta` 端点

````python
from openai import OpenAI
import os
from dotenv import load_dotenv

load_dotenv(dotenv_path='../../.env')

api_key = os.getenv("VITE_DEEPSEEK_API_KEY")
if not api_key:
    raise ValueError("❌ 未找到 API Key")

# ⚠️ 使用 beta 端点
client = OpenAI(
    api_key=api_key,
    base_url="https://api.deepseek.com/beta",
)
````

## 强制输出 Python 代码

````python
messages = [
    {"role": "user", "content": "Please write quick sort code"},
    {
        "role": "assistant",
        "content": "```python\n",
        "prefix": True  # 标记为前缀
    }
]

response = client.chat.completions.create(
    model="deepseek-chat",
    messages=messages,
    stop=["```"],  # 在代码块结束符处停止
)

print(response.choices[0].message.content)
````

## JSON 格式引导

````python
messages = [
    {"role": "user", "content": "提取人名和年龄：张三，25岁"},
    {
        "role": "assistant",
        "content": '{"name": "',
        "prefix": True
    }
]

response = client.chat.completions.create(
    model="deepseek-chat",
    messages=messages,
    temperature=0.1
)

print(response.choices[0].message.content)
````
